```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv('dulieuxettuyendaihoc.csv')
```

```python
print("--- 10 DÒNG ĐẦU TIÊN CỦA TẬP DỮ LIỆU ---")
display(df.head(10))
```

```python
print("\n--- 10 DÒNG CUỐI CÙNG CỦA TẬP DỮ LIỆU ---")
display(df.tail(10))
```

```python
print("--- KHẢO SÁT DỮ LIỆU THIẾU CỘT DÂN TỘC (DT) ---")
# Lập bảng tần số và tần suất (bao gồm cả giá trị thiếu NaN)
dt_tanso = df['DT'].value_counts(dropna=False)
dt_tansuất = df['DT'].value_counts(dropna=False, normalize=True) * 100

print("Bảng tần số cột DT:\n", dt_tanso)
print("\nBảng tần suất (%) cột DT:\n", dt_tansuất)
print("\nCác giá trị riêng biệt (Unique) trong cột DT:", df['DT'].unique())

# Thay thế các giá trị bị thiếu bằng giá trị 0 (Dùng phép gán chuẩn mới)
df['DT'] = df['DT'].fillna(0)
print("\n-> Kết quả: Đã xử lý thiếu cột DT bằng cách điền giá trị 0.")
```

```python
print("--- KHẢO SÁT DỮ LIỆU THIẾU BIẾN T1 ---")
# Tính giá trị trung bình (Mean) của cột T1
t1_mean = df['T1'].mean()
print(f"Giá trị trung bình (Mean) của cột T1 là: {t1_mean:.2f}")

# Thay thế dữ liệu thiếu bằng giá trị Mean (Dùng phép gán chuẩn mới)
df['T1'] = df['T1'].fillna(t1_mean)
print("-> Kết quả: Đã xử lý thiếu cột T1 bằng phương pháp Mean.")
```

```python
# Tạo danh sách tất cả các cột điểm số học tập lớp 10, 11, 12 và điểm thi đại học
cac_cot_diem = []
for nam in ['1', '2', '6']:
    for mon in ['T', 'L', 'H', 'S', 'V', 'X', 'D', 'N']:
        cac_cot_diem.append(f"{mon}{nam}")
cac_cot_diem.extend(['DH1', 'DH2', 'DH3'])

# Dùng vòng lặp quét qua và điền Mean (Dùng phép gán chuẩn mới)
for cot in cac_cot_diem:
    if df[cot].isnull().sum() > 0:
        mean_val = df[cot].mean()
        df[cot] = df[cot].fillna(mean_val)
        
print("-> Kết quả: Đã tự động xử lý toàn bộ dữ liệu điểm số còn thiếu bằng phương pháp Mean.")
```

```python
# Công thức: TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10
df['TBM3'] = (df['T6']*2 + df['L6'] + df['H6'] + df['S6'] + df['V6']*2 + df['X6'] + df['D6'] + df['N6']) / 10

print("-> Kết quả: Đã tạo xong các biến trung bình môn TBM1, TBM2, TBM3.")
display(df[['TBM1', 'TBM2', 'TBM3']].head())
```

```python
def dinh_dang_xep_loai(tbm):
    if tbm < 5.0: return 'Y'
    elif tbm < 6.5: return 'TB'
    elif tbm < 8.0: return 'K'
    elif tbm < 9.0: return 'G'
    else: return 'XS'

df['XL1'] = df['TBM1'].apply(dinh_dang_xep_loai)
df['XL2'] = df['TBM2'].apply(dinh_dang_xep_loai)
df['XL3'] = df['TBM3'].apply(dinh_dang_xep_loai)

print("-> Kết quả: Đã xếp loại học lực xong (XL1, XL2, XL3).")
display(df[['TBM1', 'XL1', 'TBM2', 'XL2', 'TBM3', 'XL3']].head())
```

```python
# Chuẩn hóa quy đổi tuyến tính hệ 10 sang hệ 4: US_TBM = (TBM / 10) * 4
df['US_TBM1'] = (df['TBM1'] / 10) * 4
df['US_TBM2'] = (df['TBM2'] / 10) * 4
df['US_TBM3'] = (df['TBM3'] / 10) * 4

print("-> Kết quả: Đã chuyển đổi điểm sang hệ 4 Mỹ (US_TBM1, US_TBM2, US_TBM3).")
display(df[['TBM1', 'US_TBM1', 'TBM2', 'US_TBM2', 'TBM3', 'US_TBM3']].head())
```

```python
def tinh_ket_qua_xet_tuyen(row):
    khoi_thi = row['KT']
    dh1 = row['DH1']
    dh2 = row['DH2']
    dh3 = row['DH3']
    
    if khoi_thi in ['A', 'A1']:
        diem_xet_tuyen = (dh1 * 2 + dh2 + dh3) / 4
    elif khoi_thi == 'B':
        diem_xet_tuyen = (dh1 + dh2 * 2 + dh3) / 4
    else:
        diem_xet_tuyen = (dh1 + dh2 + dh3) / 3
        
    return 1 if diem_xet_tuyen >= 5.0 else 0

df['KQXT'] = df.apply(tinh_ket_qua_xet_tuyen, axis=1)

print("--- THỐNG KÊ KẾT QUẢ XÉT TUYỂN (1: ĐẬU, 0: RỚT) ---")
print(df['KQXT'].value_counts())
```

```python
df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False)
print("[THÀNH CÔNG] Đã lưu file sạch tại thư mục hiện hành với tên: processed_dulieuxettuyendaihoc.csv")
```
